```python
from PytorchProject.main import *
from PytorchProject.utils import *
from PytorchProject.models.resnet import *
import argparse

# Take inputs from user w.r.t batch size, scheduler, learning rate etc.
parser = argparse.ArgumentParser(description='PyTorch CIFAR10 Training')
parser.add_argument('-lr', '--lr', default=0.1, type=float, help='learning rate')
parser.add_argument('-opt', '--opt', type=str, help='optimizer name')
parser.add_argument('-sch', '--sch', type=str, help='schedular name')
parser.add_argument('-bs', '--bs', type=int, help='size of batch')
args = parser.parse_args(['-lr', '0.1', '-opt', 'SGD','-sch', 'CosineAnnealingLR', '-bs', '128'])
args_dict = vars(args)
arg1 = args_dict['lr'] 
arg2 = args_dict['opt']
arg3 = args_dict['sch']
arg4 = args_dict['bs']

sample = network_sample(arg1)  # create instance of network class
model = sample.model(arg2, arg3)  # create network by calling model function
data_utils_instance=data_utils()  # create instance of utils class
train_loader,test_loader =  data_utils_instance.data('CIFAR10', arg4)  # create data loaders
train_loss_graph, train_accuracy_graph, test_loss_graph, test_accuracy_graph,  misc_im,misc_tr,misc_pred = sample.run_samples(20, train_loader, test_loader)  # run the network and store params requured for graphs
data_utils_instance.plot_accuracy_loss_graph(train_loss_graph, train_accuracy_graph, test_loss_graph, test_accuracy_graph)  # plot accuracy and loss graph
data_utils_instance = data_utils_instance.show_misclassified_images(misc_im,misc_tr,misc_pred)  # show misclassified images
```
